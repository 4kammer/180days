```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot

df = pd.read_csv("LADWP_JTI.csv") #Reading the dataset in a dataframe using Pandas
```

```python
df.head(10)
```

```python
df.describe()
```

```python
fig = plt.pyplot.figure()
ax = fig.add_subplot(111)
ax.hist(df['Amount'], bins = 10, range = (df['Amount'].min(),df['Amount'].max()))
plt.pyplot.title('Amount distribution')
plt.pyplot.xlabel('Amount')
plt.pyplot.ylabel('Count of Amount')
plt.pyplot.show()
```
