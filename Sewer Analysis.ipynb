```python
import graphlab
```

```python
data = graphlab.SFrame.read_csv('desktop/sewerdata.csv', na_values=["NA"])
```

```python
data = data.dropna(['ASSETID','PIPE_ID','PIPE_SUBTYPE','UP_STRUCT','DN_STRUCT','PIPE_LABEL','REMARKS','STREET','BLOCK','LINER'])
```

```python
data.head
```

```python
graphlab.canvas.set_target('ipynb')
data.show()
```

```python
# split the data randomly, keeping 80% for training and the rest for validation
(train, test) = data.random_split(0.8)
```

```python
model = graphlab.boosted_trees_classifier.create(train, target='PIPE_STATUS')
```

```python
predictions = model.classify(test)
```

```python
results = model.evaluate(test)
```
